### Matplotlib 的基础使用之多个子图

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

# jupyter专用，可以不使用plt.show()
%matplotlib inline

# 设置字体为支持中文的字体 (matplotlib默认是不支持中文的)
mpl.rcParams['font.family'] = 'SimHei'  # 使用黑体
mpl.rcParams['axes.unicode_minus'] = False  # 正确显示负号
```

```python
# 数据准备
x1 = np.linspace(0.0, 5.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)

x2 = np.linspace(0.0, 2.0)
y2 = np.cos(2 * np.pi * x2)
```

```python
# 绘制简单的子图
# 两个子图，放一列
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('一张图里面放两张子图')

ax1.plot(x1, y1, 'o-')
ax1.set_ylabel('Damped oscillation')

ax2.plot(x2, y2, '.-')
ax2.set_xlabel('time (s)')
ax2.set_ylabel('Undamped')

# 以下可以注销，然后第一行的ax3删去，子图数量改成2, 1即可
# ax3.plot(x1, y2, '.-')
# ax3.set_xlabel('time (s)')
# ax3.set_ylabel('Undamped')

plt.show()
```
