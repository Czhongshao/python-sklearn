```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set(palette='deep')
```

```python
tips = sns.load_dataset('tips',data_home='seaborn-data',cache=True)
g = sns.jointplot(x="total_bill", y="tip", data=tips, color="darkblue")
```

```python
sns.set_theme()
g = sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg", color="darkblue")
```

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
sns.set_theme()
# sns.set_palette(sns.light_palette("navy"))
g = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex", color="darkblue")
```

```python
iris = sns.load_dataset("iris")
g = sns.jointplot(x="sepal_width", y="petal_length", data=iris,kind="kde", space=0, color="g")
```

```python
g = sns.jointplot(x="sepal_length", y="sepal_width", data=iris, color="k")
g.plot_joint(sns.kdeplot, zorder=0, n_levels=6)
```
