### 逻辑回归算法
#### 逻辑回归本质上是一个分类方法，主要用于二分类的问题(输出只有两种，分别代表两个类别)
##### Logisitc回归利用了Logistic函数(或称 Sigmoid函数)，也称Logistic回归分析

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
    # 直接返回Sigmoid函数
    return 1 / (1 + np.exp(-x))
```

```python
def plot_sigmoid():
    # param: 起点，终点，间距
    x = np.arange(-8, 8, 0.1)
    y = sigmoid(x)
    plt.plot(x, y)
    plt.show()
```

```python
if __name__ == '__main__':
    plot_sigmoid()
    
```
