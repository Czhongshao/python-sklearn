### 利用seaborn 进行图像绘制

#### 案例1

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

##### 初始化参数

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
sns.set()
sinplot()
```

#### **五种主题风格**
 - darkgrid
 - whitegrid
 - dark
 - white
 - ticks

```python
sns.set_style("ticks")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style("darkgrid")
sinplot()
```

```python
sns.set_style("whitegrid")
sinplot()
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

```python
# 去除某些轴
sinplot()
sns.despine()
```

```python
#f, ax = plt.subplots()
sns.violinplot(data)
sns.despine(offset=10)
```

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
# 隐藏左轴
sns.despine(left=True)
```

```python
# with 执行某种风格，在其中的都会执行该风格的内容
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```python
sns.set() # 设置默认值
```
