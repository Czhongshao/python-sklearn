```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

# jupyter专用，可以不使用plt.show()
%matplotlib inline

# 设置字体为支持中文的字体 (matplotlib默认是不支持中文的)
mpl.rcParams['font.family'] = 'SimHei'  # 使用黑体
mpl.rcParams['axes.unicode_minus'] = False  # 正确显示负号

# 数据准备
x1 = np.linspace(0.0, 5.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)

x2 = np.linspace(0.0, 2.0)
y2 = np.cos(2 * np.pi * x2)

# 创建子图框架
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('一张图里面放两张子图')

# 在第一个子图上绘制
sns.lineplot(x=x1, y=y1, ax=ax1, marker='o')
ax1.set_ylabel('Damped oscillation')

# 在第二个子图上绘制
sns.lineplot(x=x2, y=y2, ax=ax2, marker='.')
ax2.set_xlabel('time (s)')
ax2.set_ylabel('Undamped')

plt.show()
```
