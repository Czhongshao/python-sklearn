### 决策树最重要的是选择特征，这个选择特征的标准就是依靠熵。因此，在学习决策树之前，需要先了解决策树中的熵，包括信息熵、条件熵、联合熵、互信息

### 1.信息熵

##### 在了解信息熵之前，需要知道信息量。信息量是通过概率来定义的：如果某个时间发生的而概率很低，那么它的信息量就很大；反之，如果某个事件发生的概率很高，那么它的信息量就很小。简而言之，概率低的事件信息量大，因此信息量可以由公式计算。

#### 需要注意，并不是概率越高的事件，信息熵越高，因为信息量与概率成反比，发生概率越高的事件，其携带的信息量越低。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eps = 1e-4
p = np.linspace(eps, 1-eps, 100)
h = -(1 - p) * np.log2(1 - p) - p * np.log2(p)
plt.plot(p, h)
plt.show()
print(max(h))
print(list(h).index(max(h)))
print(p[list(h).index(max(h))])
```

#### 当某个事件发生的概率为 0.5050494949494949 时，信息熵有最大值 0.999926428806495 。
##### 例如：NBA 季后赛两只球队的获胜概率都是0.5，那么会有更多的人去观看，这次球赛的激烈程度是最大的，带来的信息量也是非常庞大的。

### 4.互信息
##### 互信息用于判断两个变量之间是否存在关系，是信息论中一种有用的信息度量，它可以看成是一个随机变量中包含的关于另一个随机变量的信息量，因此互信息也被称为信息增益。
