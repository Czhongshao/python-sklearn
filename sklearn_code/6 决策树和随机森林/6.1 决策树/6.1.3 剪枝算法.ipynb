### 剪枝算法
##### 为了解决决策树的过拟合问题，经常需要对决策树进行剪枝处理，将分的过细的叶子节点删除，回退到其父节点或更高的节点，使其父节点或更高的节点变为叶子节点。
##### 决策树的剪枝策略最基本的有两种：预剪枝和后剪枝。

#### 预剪枝
##### 在决策树生成过程中进行剪枝，在每次生成决策树叶子节点时，通过一些常见的方法考虑是否能够带来决策树性能的提升。
##### 预剪枝方法主要有如下几种：当数的深度达到一定规模时，决策树停止生长；当达到叶子节点的数量大于某个阈值时，决策树停止生长；当信息增益和信息增益率和基尼指数增益小于某个阈值时，决策树不再生长。
##### 预剪枝在实际中的效果并不好。

#### 后剪枝
##### 是目前剪枝算法中最常见、用的最多的做法，它是在决策树生成完成后，再进行剪枝。
##### 剪枝的过程是对拥有父节点和叶子节点的字数节点进行检查。
#### 判断条件
##### 将字数叶子节点删除，判断删除前后的错分率，如果错分率在删除后变小，则将字数节点中的叶子节点删除。
#### 常见的后剪枝方法：错误率降低剪枝、悲观错误剪枝和代价复杂度剪枝

### 1.错误率降低剪枝
#### 错误率降低剪枝（REP）是通过一个新的验证集来纠正树的过拟合问题，是最简单的剪枝方法
##### 首先用训练样本生成决策树后，自上而下对于每个节点决定是否修剪该节点。先假设要删除该节点下的子树节点使其成为叶子节点，再用验证样本比较修建前后的错分率，当修剪后的错分率比修建前的小时，便删除该节点下的叶子节点。

### 2.悲观错误剪枝
#### 悲观错误剪枝（PEP）是在C4.5决策树算法中提出的
##### 悲观错误剪枝不需要想错误率降低剪枝用部分样本作为测试数据，而是完全使用训练集数据来生成决策树，然后再用这些训练集数据来完成剪枝操作。

### 3.代价复杂度剪枝
#### 代价复杂度剪枝（CCP）主要用于CART算法。
##### 代价复杂度剪枝为子树T定义了代价和复杂度，以及一个可由用户设置的衡量代价与复杂度之间的关系参数α。
